### 1. Import Libraries

```python
import pandas as pd
import numpy as np
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

### 2 Loading Data

```python
np.random.seed(7)
data= np.loadtxt("pima-indians-diabetes.csv",delimiter=",")
print(data)
len(data)
```

### 3. Defining Models

```python
def build_models():
    model=Sequential()
    model.add(Dense(12,input_dim=8,activation='relu'))
    model.add(Dense(8,activation='relu'))
    model.add(Dense(1,activation='sigmoid'))
    model.compile(loss='binary_crossentropy',optimizer ='adam',metrics=['accuracy'])
    return model
```

### 4. Initializing epoch / Preparing data


```python
k=4
num_validation_samples=len(data)//k
np.random.shuffle(data)
validation_scores=[]
```

#### 5 For loop to iterate over  k-folds

```python
for fold in range(k):
    print('Fold # ' , fold+1)
    validation_data=data[num_validation_samples*fold:num_validation_samples*(fold+1)]
    print('len of validation_data:',len(validation_data))
    training_data=np.vstack((data[:num_validation_samples*fold],data[num_validation_samples*(fold+1):]))    
    print('len of training_data:',len(training_data))
        
    X=training_data[:,0:8]
    Y_labels=training_data[:,8]
    
    model= build_models()
    model.fit(X,Y_labels,epochs=100, batch_size=15)
    
    X_val =validation_data[:,0:8]
    Y_val =validation_data[:,8]
    
    val_loss,val_accuracy=model.evaluate(X_val,Y_val)
    print('val_loss:',val_loss)
    print('val_accuracy:',val_accuracy)
    
    print('X_Train:',len(X))
    print('Y_label_Train:',len(Y_labels))
    
    validation_scores.append(val_accuracy)
    
```
